1. Artifact Representation

```python
class Artifact:
    def __init__(self, data):
        self.title = data['title']
        self.date = data['creation_date']
        self.url = data['url']
        self.creators = [c['description'] for c in data['creators'] if 'description' in c]  # Extract only 'description' within the 'creators' attribute

    def __str__(self): # string representation
        return f"title: {self.title}\n creators:{self.creators}"

    def to_rdf(self):

        EX = Namespace("http://w3id.org/example/")
        CRM = Namespace("http://www.cidoc-crm.org/cidoc-crm/")
        DC = Namespace("http://purl.org/dc/elements/1.1/")

        graph = rdflib.Graph()

        graph.bind("ex", EX)
        graph.bind("crm", CRM)
        graph.bind("dc", DC)

        artifact_uri = rdflib.URIRef(f"http://example.org/artifact/{self.title.replace(' ', '_')}")

        # Add triples to the graph
        graph.add((artifact_uri, rdflib.RDF.type, rdflib.URIRef("http://example.org/Artifact"))) 
        graph.add((artifact_uri, rdflib.URIRef("http://example.org/title"), rdflib.Literal(self.title)))
        graph.add((artifact_uri, rdflib.URIRef("http://example.org/creation_date"), rdflib.Literal(self.date)))
        graph.add((artifact_uri, rdflib.URIRef("http://example.org/url"), rdflib.Literal(self.url)))
        
        for creator in self.creators:
            creator_uri = rdflib.URIRef(f"http://example.org/creator/{creator.replace(' ', '_')}")
            graph.add((artifact_uri, rdflib.URIRef("http://example.org/creator"), creator_uri))
            graph.add((creator_uri, rdflib.RDF.type, rdflib.URIRef("http://example.org/Creator")))
            graph.add((creator_uri, rdflib.URIRef("http://example.org/name"), rdflib.Literal(creator)))

        return graph

    def similar_wikidata_enrich(self):
        sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
    
        safe_title = self.title.replace('"', '\\"')  # Error handling for HTTP errors
        
        query = f"""
        SELECT ?item ?itemLabel
        WHERE {{
        ?item wdt:P31 wd:Q3305213;  # Instance of artwork
                rdfs:label ?itemLabel.
        FILTER(LANG(?itemLabel) = "en")
        FILTER(CONTAINS(LCASE(?itemLabel), LCASE("{safe_title}")))
        }}
        LIMIT 5
        """

        sparql.setQuery(query)
        sparql.setReturnFormat(JSON)
        results = sparql.query().convert()

        enrich = [result["itemLabel"]["value"] for result in results["results"]["bindings"]]
        return enrich


    def artworks(self):
        # query both Chicago or Cleveland API for similar artworks
        chicago_api_url = f"https://api.artic.edu/api/v1/artworks/search?q={self.title}"
        response = requests.get(chicago_api_url)
        data = response.json()

        # Extract artwork information
        similar_artworks = []
        for artwork in data['data']:
            print(artwork)
            artwork_title = artwork['title']
            #artwork_creator = artwork['artist_display']
            #similar_artworks.append({"title": artwork_title})#, "creator": artwork_creator})
            similar_artworks.append(artwork_title)

        return similar_artworks
```

```python
from rdflib import Graph, URIRef, Literal
from rdflib.namespace import RDFS, XSD
from rdflib.namespace import DC, DCTERMS, DOAP, FOAF, SKOS, OWL, RDF, RDFS, VOID, XMLNS, XSD
```

```python
class Artist:
    def __init__(self,creator):

        #print(creator['description'])
        line = creator['description']
        tokens = line.split("(")  
        name = tokens[0].strip()
        nationality = "Not Found"
        if len(tokens)>1:
            tmp = tokens[1]
            tmp2 = tmp.split(",")
            nationality = tmp2[0].strip()
        self.name = name
        self.nationality = nationality
        self.artifacts = []  
    def __str__(self):
        return f"{self.name}"

    # simplistic proof of concept
    def to_rdf(self):
        EX = Namespace("http://example.org/")
        g = Graph()
        g.bind("ex", EX)
        g.bind("foaf", FOAF)

        artist_uri = URIRef(f"http://example.org/creator/{self.name.replace(' ', '_')}")

        g.add((artist_uri, RDF.type, EX.Creator))
        g.add((artist_uri, FOAF.name, Literal(self.name)))
        g.add((artist_uri, EX.nationality, Literal(self.nationality)))

        # struture artist-artifacts relationship 
        for title in self.artifacts:
            artifact_uri = URIRef(f"http://example.org/artifact/{title.replace(' ', '_')}")
            g.add((artifact_uri, RDF.type, EX.Artifact))
            g.add((artist_uri, EX.creatorOf, artifact_uri))
            g.add((artifact_uri, EX.title, Literal(title)))

        return g


def wikidata_enrichment(self):
    sparql = SPARQLWrapper("https://query.wikidata.org/sparql")

    safe_name = self.name.replace('"', '\\"')

    # wiki query to find e.g. the artist's genre
    query = f"""
    SELECT ?genre ?genreLabel
    WHERE {{
        ?artist rdfs:label "{safe_name}"@en;  # Search for artist by name
                wdt:P31 wd:Q5.  # Instance of human (artist)
        ?artist wdt:P136 ?genre.  # Artist's genre (P136)
        ?genre rdfs:label ?genreLabel.  # Genre's label
        FILTER(LANG(?genreLabel) = "en")
    }}
    LIMIT 1
    """

    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()

    if results["results"]["bindings"]:
        # Extract the genre label
        genre = results["results"]["bindings"][0]["genreLabel"]["value"]
        self.genre = genre  # Enrich artist with the genre
        return genre
    else:
        return None
```

```python
import rdflib
from rdflib import Graph, Literal, RDF, URIRef, Namespace
from rdflib.namespace import RDFS
import matplotlib.pyplot as plt
from collections import Counter
import requests
import os

# week 5 Übung
# create a namespace object
CRM = Namespace("http://www.cidoc-crm.org/cidoc-crm/") #CRM is an ontology used in museums, archives and CH institutions to describe objects, events and people and their relation
DC = Namespace("http://purl.org/dc/elements/1.1/") # for simple metadata like title, name 
EX = Namespace("http://w3id.org/example/") #custom namespace

class Collection:
    def __init__(self):
        self.artifacts = []
        self.artists = {}

    def add_artifact(self, artifact):
        self.artifacts.append(artifact)
        if artifact.artist and artifact.artist.name not in self.artists:
            self.add_artist(artifact.artist)

    def add_artist(self, artist):
        self.artists[artist.name] = artist

    
    def to_rdf(self, output_path="RDF_outputs/collection.ttl"):

        g = Graph()
    
        g.bind("crm", CRM)
        g.bind("dc", DC)
        g.bind("ex", EX)
        g.bind("rdfs", Namespace("http://www.w3.org/2000/01/rdf-schema#"))

        for artifact in self.artifacts:
            artifact_uri = EX[f"artifact{artifact.id:03d}"]
            artist_uri = EX[f"artist{artifact.id:03d}"]
            museum_uri = EX[f"museum{artifact.id:03d}"]

    # Add triples for the Artifact
            g.add((artifact_uri, RDF.type, CRM["E22_Man-Made_Object"]))
            g.add((artifact_uri, DC.title, Literal(artifact.title)))
            g.add((artifact_uri, DC.creator, artist_uri))
        
            if artifact.museum:
                # Link artifact to the current museum / place
                g.add((artifact_uri, CRM["P55_has_current_location"], museum_uri))
        
    # Add triples for the Artist
            g.add((artist_uri, RDF.type, CRM["E21_Person"]))
            g.add((artist_uri, RDFS.label, Literal(artifact.artist.name)))

    # Add triples for the Museum if present
            g.add((museum_uri, RDF.type, CRM["E53_Place"]))
            g.add((museum_uri, RDFS.label, Literal(artifact.museum)))

        os.makedirs(os.path.dirname(output_path), exist_ok=True)
        g.serialize(destination=output_path, format="turtle") # serialize the graph to Turtle ttl format
        print(f" RDF file was created: {output_path}")
        return g

    def visualize_metadata(self): 
        artist_counts = Counter([artifact.artist.name for artifact in self.artifacts if artifact.artist])

        artists = list(artist_counts.keys())
        counts = list(artist_counts.values())

        plt.figure(figsize=(8, 5))
        plt.bar(artists, counts, color='skyblue') # visualize it in bar chart
        plt.xlabel("Artists")
        plt.ylabel("Number of Artifacts")
        plt.title("Number of Artifacts per Artist")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    def visualize_rdf(self):
        import networkx as nx

        g = self.to_rdf()
        nx_graph = nx.DiGraph()

        for s, p, o in g:
            nx_graph.add_edge(str(s), str(o), label=str(p))

        pos = nx.spring_layout(nx_graph)
        plt.figure(figsize=(10, 8))
        nx.draw(nx_graph, pos, with_labels=True, node_size=2000, node_color="lightblue", arrows=True)
        edge_labels = nx.get_edge_attributes(nx_graph, 'label')
        nx.draw_networkx_edge_labels(nx_graph, pos, edge_labels=edge_labels)
        plt.title("RDF Graph representation")
        plt.show()

    def cross_api_enrich(self):
        #Enriches collection by fetching additional works by existing artists from Chicago AIC and Cleveland APIs
        base_aic = "https://api.artic.edu/api/v1/artworks/search"
        base_cleveland = "https://openaccess-api.clevelandart.org/api/artworks/"

        for artist_name in self.artists.keys():
            # AIC API
            params = {"q": artist_name, "limit": 3}
            aic_response = requests.get(base_aic, params=params).json()
            if "data" in aic_response:
                for item in aic_response["data"]:
                    artifact = Artifact(
                        id=f"aic_{item['id']}",
                        title=item.get("title", "Untitled"),
                        year=item.get("date_display", None),
                        artist=self.artists[artist_name]
                    )
                    self.add_artifact(artifact)

            # Cleveland API
            cleveland_response = requests.get(base_cleveland, params={"artist": artist_name, "limit": 3}).json()
            if "data" in cleveland_response:
                for item in cleveland_response["data"]:
                    artifact = Artifact(
                        id=f"cleveland_{item['id']}",
                        title=item.get("title", "Untitled"),
                        year=item.get("creation_date", None),
                        artist=self.artists[artist_name]
                    )
                    self.add_artifact(artifact)


# minimal Artifact and Artist classes
class Artist:
    def __init__(self, name, birth_year=None):
        self.name = name
        self.birth_year = birth_year


class Artifact:
    def __init__(self, id, title, year=None, artist=None, museum=None):
        self.id = id
        self.title = title
        self.year = year
        self.artist = artist
        self.museum = museum


collection = Collection() # erzeugt ein Objekt ,,collection" to the ,,Collection" class

for i in range(1, 51):
    artist = Artist(name=f"Artist {i}", birth_year=1800 + i)
    artifact = Artifact(
        id=i,
        title=f"Artifact number {i}",
        year=1900 + i,
        artist=artist,
        museum=f"Museum {i}"
    )
    collection.add_artifact(artifact)

# Generate RDF and save
collection.to_rdf()
```
